```python
from glob import glob
import pandas as pd

adimplentes = glob('raw/adimplentes*')
inadimplentes = glob('raw/inadimplentes*')
dim = glob('raw/relat*')
```

```python
# juntando adimplentes com concat
df = pd.DataFrame()
for file in adimplentes:
    print(file)
    df_temp = pd.read_csv(file, sep=';')
    df = pd.concat([df, df_temp])
df
```

```python
df.info()
```

```python
df['dt_vencimento'] = pd.to_datetime(df['dt_vencimento'])
df['dt_pgto'] = pd.to_datetime(df['dt_pgto'])
```

```python
df.info()
```

```python
df = df.sort_values(by='dt_vencimento').reset_index(drop=True)
data_min = df['dt_pgto'].min()
data_max = df['dt_pgto'].max()
df
```

```python
df.to_parquet('processed/fato_adimplentes.parquet', engine='pyarrow')
```

```python
# juntando inadimplentes com concat
df = pd.DataFrame()
for file in inadimplentes:
    print(file)
    df_temp = pd.read_csv(file, sep=';')
    df = pd.concat([df, df_temp])
df['dt_vencimento'] = pd.to_datetime(df['dt_vencimento'])
df = df.sort_values(by='dt_vencimento').reset_index(drop=True)
df.to_parquet('processed/fato_inadimplentes.parquet', engine='pyarrow')
```

```python
# gerando dimensões
df = pd.read_csv(dim[0], sep=';')
df.head()
```

```python
df.columns
```

```python
d_calendario = pd.DataFrame({'data': pd.date_range(data_min, data_max)})
d_calendario['ano'] = d_calendario['data'].dt.year
d_calendario['mes_num'] = d_calendario['data'].dt.month
d_calendario['mes_nome'] = d_calendario['data'].dt.month_name()
d_calendario['trimestre'] = d_calendario['data'].dt.quarter
d_calendario['dia_semana'] = d_calendario['data'].dt.day_name()
d_calendario['mes_ano_ref'] = d_calendario['data'].dt.strftime('%Y-%m')
```

```python
d_calendario
```

```python
import numpy as np
# Extrai o ano como string
ano = d_calendario['data'].dt.year.astype(str)

# Define o semestre: .1 se mês <= 6, senão .2
d_calendario['periodo_letivo'] = ano + np.where(d_calendario['data'].dt.month <= 6, '.1', '.2')
```

```python
d_calendario
```

```python
d_calendario.to_parquet('processed/dim_calendario.parquet', engine='pyarrow')
```
